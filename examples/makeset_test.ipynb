```julia
using AnomalyDetection
using StatsBase
using PyPlot
```

```julia
# set the path to Loda data
master_path = "../../../data/Loda/public/datasets/numerical"
```

```julia
# load all the Loda data into one dictionary
data = AnomalyDetection.loaddata(master_path)
```

```julia
# does it cluster well?
#testset = AnomalyDetection.Basicset(joinpath(master_path, "iris"))
testset = data["iris"]
train, test, clusterdness = AnomalyDetection.makeset(testset, 0.8, "easy", 0.05, "high")
scatter3D(train.data[1,1:80], train.data[2,1:80], train.data[3,1:80])
scatter3D(train.data[1,81:end], train.data[2,81:end], train.data[3,81:end])
title("high clusterdness = $(clusterdness)")
```

```julia
train, test, clusterdness = AnomalyDetection.makeset(testset, 0.8, "easy", 0.05, "low")
scatter3D(train.data[1,1:80], train.data[2,1:80], train.data[3,1:80])
scatter3D(train.data[1,81:end], train.data[2,81:end], train.data[3,81:end])
title("low clusterdness = $(clusterdness)")
```

```julia
# does it cluster well?
#testset = AnomalyDetection.Basicset(joinpath(master_path, "iris"))
testset = data["iris"]
train, test, clusterdness = AnomalyDetection.makeset(testset, 0.8)
nn = length(train.labels[train.labels .== 0])
scatter3D(train.data[1,1:nn], train.data[2,1:nn], train.data[3,1:nn])
scatter3D(train.data[1,nn+1:end], train.data[2,nn+1:end], train.data[3,nn+1:end])
title("high clusterdness = $(clusterdness)")
```

```julia
nn = length(test.labels[test.labels .== 0])
scatter3D(test.data[1,1:nn], test.data[2,1:nn], test.data[3,1:nn])
scatter3D(test.data[1,nn+1:end], test.data[2,nn+1:end], test.data[3,nn+1:end])
title("high clusterdness = $(clusterdness)")
```
