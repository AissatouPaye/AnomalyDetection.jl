```julia
# set the path to Loda data
master_path = "../../../data/Loda/public/datasets/numerical"
```

```julia
# load all the Loda data into one dictionary
data = AnomalyDetection.loaddata(master_path)
```

```julia
# does it cluster well?
#testset = AnomalyDetection.Basicset(joinpath(master_path, "iris"))
testset = data["iris"]
x, y, z = AnomalyDetection.makeset(testset, 0.8, "easy", 0.05, "high")
scatter3D(x.data[1,1:80], x.data[2,1:80], x.data[3,1:80])
scatter3D(x.data[1,81:end], x.data[2,81:end], x.data[3,81:end])
title("high clusterdness = $(z)")
```
